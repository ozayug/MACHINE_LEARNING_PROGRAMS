```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('stock_data.csv', parse_dates=['Date'])
```

```python
df
```

```python
start_date = '2024-01-02'
end_date = '2024-01-06'
```

```python
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
```

```python
filtered_df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Opening Price', color='blue', marker='o')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Closing Price', color='red', marker='o')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Opening and Closing Stock Prices')
plt.legend()
plt.grid(True)
plt.show()
```
